```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('task5_1.csv')
df
```

```python
fig = plt.figure(figsize=(8, 6))
bars = plt.bar(df['date'], df['average_request_to_pickup'])

# Bar annotation
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(int(height)), ha='center', va='bottom', fontsize=8)

plt.xlabel('January Days')
plt.ylabel('Average waiting time (s)')
plt.title('Average waiting time on each day in January (s)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
fig.savefig('average_waiting_time.png')
```
