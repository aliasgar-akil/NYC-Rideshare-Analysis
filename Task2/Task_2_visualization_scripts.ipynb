## 1. Histogram of number of trips per business in each month

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('task2_1.csv')
df
```

```python
df['date'] = df['date'].apply(lambda x: x[6])
df['business'] = df['business'] + '-' + df['date']
df.drop('date', axis=1, inplace=True)
df
```

```python
fig = plt.figure(figsize=(8, 6))
bars = plt.bar(df['business'], df['count'])

# Bar annotation
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(int(height)), ha='center', va='bottom', fontsize=8)

plt.xlabel('Business-month')
plt.ylabel('Number of trips')
plt.title('Number of trips per business per month')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
fig.savefig('business_count_hist.png')
```
